#### Making First Visual: Max Temperature per Month

```python
# Change date type
df2['date'] = pd.to_datetime (df2['date'])
```

```python
#check data type changed
df2.info()
```

```python
# add a month and year column
df2['month'] = df2['date'].dt.month
df2['year'] = df2['date'].dt.year
```

```python
#check for added columns
df2.info()
```

```python
# drop null values
fil = df2[df2['max_temp_c'] != '\\N']
```

```python
# conver max_temp_c to float data type
fil['max_temp_c'] = fil['max_temp_c'].astype(float)
```

```python
# checking for unique year values
fil['year'].unique()
```

```python
# dataframe with only year, month, and temp (sorting too)
filx = fil[['year', 'month', 'max_temp_c']].sort_values(by=['year', 'month', 'max_temp_c'])
```

```python
# checking sort
filx.head(5)
```

```python
#checking data types
filx.info()
```

```python
# make new data frame with month names associated with max temp per month
data = []
months = {1: 'January', 2: 'February', 3: 'March', 4: 'April', 5: 'May', 6: 'June',
         7: 'July', 8: 'August', 9: 'September', 10: 'October', 11: 'November', 12: 'December'}
for i in range(1900, 2024):
    for j in range(1, 13):
        max_temp = filx[(filx['year'] == i) & (filx['month'] == j)]['max_temp_c'].max()
        row = {'year': i, 'month': months[j], 'max_temp_c': max_temp}
        data.append(row)

vdf = pd.DataFrame(data)
print(vdf)
```

```python
# remove null values
vdf = vdf.dropna()
```

```python
# define the slider object
slider = alt.binding_range(min = 1900, max = 2023, step = 1, name='year')
op_var = alt.param(value= 1900, bind=slider)

# create the scatter plot with the slider
alt.Chart(vdf).mark_bar(color = 'orange').encode(
    x = alt.X('month', sort=["January", "February"], title='Month'),
    y = alt.Y('max_temp_c:Q', title='Maximum Temperature (C)'),
    tooltip = 'max_temp_c'
).properties(
    title = 'World Wide Maximum Temperature per Month 1900-2023',
    width = 600,
    height = 400
).add_params (
    op_var
).transform_filter (
    alt.datum.year == op_var
)
```

#### Third Visual: CO2, Temp, and Disaster Count

```python
df1.rename(columns={'Year': 'year'}, inplace=True)
```

```python
series1 = df1.groupby('year')['Disaster Type'].count()
series1
```

```python
series2 = df3.groupby('year')['value'].mean()
series2
```

```python
# Drop rows where either column has '\N'
df2_clean = df2[(df2['avg_temp_c'] != '\\N') & (df2['snow_depth_mm'] != '\\N')].copy()

df2_clean['avg_temp_c'] = df2_clean['avg_temp_c'].astype(float)
df2_clean['snow_depth_mm'] = df2_clean['snow_depth_mm'].astype(float)
```

```python
df2_clean.info()
```

```python
series3 = df2_clean.groupby('year')['avg_temp_c'].mean()
series3
```

```python
merged_series = pd.merge(series2, series3, on='year', how='inner')
merged_series
```

```python
final_merge = pd.merge(merged_series, series1, on='year', how='inner')
final_merge = final_merge.reset_index()
```

```python
alt.Chart(final_merge).mark_circle(color='blue').encode(
    alt.X('year:N', title = 'Year').scale(zero=False),
    alt.Y('value', title = 'CO2 Levels').scale(zero=False, padding=1),
    size=alt.Size('Disaster Type', title='Num of Disasters'),
    tooltip=[
        alt.Tooltip('year', title='Year'),
        alt.Tooltip('value', title='CO2 Levels'),
        alt.Tooltip('Disaster Type', title='Num of Disasters')
    ]
)
```

#### Fourth Visual

```python
df2_clean.info()
```

```python
four = df2_clean[['year', 'city_name', 'snow_depth_mm', 'avg_temp_c']]
four
```

```python
def classify_climate(temp):
    if temp >= 20:  # threshold 
        return 'Warm'
    else:
        return 'Cold'
        
four['climate_type'] = four['avg_temp_c'].apply(classify_climate)
four
```

```python
fourth = four.groupby(['year', 'climate_type'])[['avg_temp_c', 'snow_depth_mm']].mean()
fourth = fourth.reset_index()
fourth
```

```python
selection = alt.selection_point(fields=['climate_type'], bind='legend')

# Create the chart with selection
alt.Chart(fourth).mark_line().encode(
    x=alt.X('year:Q', title='Year'),
    y=alt.Y('snow_depth_mm:Q', title='Snow Depth (mm)'), 
    color=alt.Color('climate_type:N', legend=alt.Legend(title='Climate Type')),
    tooltip='snow_depth_mm:Q',
).properties(
    title='Average Seasonal Snowfall'
)
```
